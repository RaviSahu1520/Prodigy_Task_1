Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

```python
#Importing Necessary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('population.csv' , index_col=['Country Name', 'Country Code', 'Indicator Name' ,'Indicator Code', ])
```

```python
# Choose Some countries for comparison
selected_countries = ["India", "China", "Afghanistan","Pakistan", "Argentina"]
population_data = df.loc[selected_countries, :]
```

```python
# Extract years and corresponding population data for each country
years = population_data.columns[2:]
population_values = population_data.iloc[:, 2:]
```

```python
# Assuming population_values is a DataFrame with country names as columns
countries = population_values.columns
```

```python
# Plotting the bar chart Countrywise
plt.figure(figsize=(18, 5))
population_values.T.plot(kind='bar', ax=plt.gca(), color=['blue', 'yellow', 'red', 'green', 'pink'])
plt.xlabel('Year Wise')
plt.ylabel('Population')
plt.title('Population Bar Chart Comparison With other Country ')
plt.legend(title='Country')
```

```python
# Plotting the histogram
plt.figure(figsize=(18, 5))
plt.subplot(1, 1, 1)
population_values.T.plot(kind='hist', bins=5, alpha=0.7, stacked=True, ax=plt.gca(), color=['blue', 'yellow', 'red', 'green', 'pink'])
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title('Population Histogram Comparison')
plt.legend(title='Country')
```

```python
# Assuming population_values is a DataFrame with years as index and population values as columns
years = population_values.index
population = population_values.iloc[:, 2].values
# Plotting the Pie
plt.figure(figsize=(12, 7))
plt.pie(population, labels=years, autopct='%1.1f%%', colors=['blue', 'yellow', 'red', 'green', 'pink'])
plt.title('Population Distribution by Year')
plt.show()
```
